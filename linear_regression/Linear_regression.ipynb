```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
X = []
Y = []
```

```python
for line in open('data_1d.csv'):
    x,y = line.split(',')
    X.append(float(x))
    Y.append(float(y))
    
```

```python
X = np.array(X)
Y = np.array(Y)
```

```python
plt.scatter(X,Y)
plt.title('X,Y data')
plt.show
```

```python
denominator = X.dot(X) - X.mean() * X.sum()
print(denominator)
```

```python
a = (X.dot(Y) - Y.mean() * X.sum()) / denominator
print(a)
```

```python
b = (Y.mean()* X.dot(X) - X.mean() * X.dot(Y)) / denominator
print(b)
```

```python
Yhat = a*X + b
```

```python
plt.scatter(X,Y)
plt.plot(X, Yhat, 'r')
plt.title('Linear Regression Line. Slope is: ' + str(round(a)) + '. Intercept: ' + str(round(b)))
plt.show()
```
